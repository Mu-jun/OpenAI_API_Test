```python
# system_content
system_content = """너는 영어 선생님으로서 중학교 1학년  학생의 영어 학습에  필요한 시나리오를 [생성조건]에 제시된 항목들을 참고하여 작성해줘.
[작성형식:json]
{
"title": "",
"Characters":[""],
"dialogue": [
{
"speaker": "",
"speech": "",
"image_prompt":""
}
]
}
- [title] : 시나리오에 어울리는 적절한 타이틀 제시 
- [dialogue]
  a. [대사]에 맞는 [이미지 프롬프트]를 작성.
  b. 제시된 [단어]를 하나씩 사용하여  설정된 [인물]이 말하는 [대사]를 영어로 작성 .
  c. 주어진 [언어형식]을 활용하고, [필수표현]이 반드시 포함될 수 있도록할 것.
"""

# user_content
material = "내가족과 친구"
context = "사람에 대해 묻고 답하기"
language_form = "be 동사의 긍정문/부정문,일반동사의 긍정문/부정문"
required_expression = """- Who is ~? 
- What is he(she) like? 
- How do you like (somebody)?"""
words = "about, actually, all day long, basketball, be from, both, chocolate"""

user_content = f"""[Creation Conditions].
1. 소재: {material}
2. 상황: {context}
3. 언어형식 : {language_form}
4. 필수표현 : {required_expression}
5. 단어 : {words}"""

completion = openai.ChatCompletion.create(
  model="gpt-3.5-turbo-16k-0613",
  messages=[
    {"role": "system", "content": system_content},
    {"role": "user", "content": user_content}
  ]
)
```

```python
print(completion.choices[0].message.content)
```
