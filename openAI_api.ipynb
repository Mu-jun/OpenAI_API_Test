```python
# system_content
system_content = """As an English teacher, please write a scenario necessary for learning English for a first-year middle school student by referring to the items presented in [Generation conditions].
[Type: json]
{
"title": "",
"Character":[""],
"dialogue": [
{
"speaker": "",
"speech": "",
"image_prompt":""
}
]
}
- [title]: Present appropriate title for the scenario
- [dialogue]
a. Generate [image_prompt] for each [Dialogue] that can effectively describe
b. Write the [speech] in English that the [Character] has set using the [words] presented one by one by one.
c. Utilize the given [language form] and ensure that [essential expressions] are included.
"""

# user_content
topic = "My family and friends"
situation = "Ask and answer about people"
language_form = "positive/negative sentences of be verbs, positive/negative sentences of general verbs"
essential_expressions = """- Who is ~?
- - What is he(she) like?
- - How do you like (somebody)?"""
words = "about, actually, all day long, basketball, be from, both, chocolate"""

user_content = f"""[Generation conditions].
1. [Topic]: {topic}
2. [Situation]: {situation}
3. [Language form]: {language_form}
4. [Essential expressions]: {essential_expressions}
5. [Words]: {words}"""

completion = openai.ChatCompletion.create(
  model="gpt-3.5-turbo-16k-0613",
  messages=[
    {"role": "system", "content": system_content},
    {"role": "user", "content": user_content}
  ]
)
```

```python
print(completion.choices[0].message.content)
```
